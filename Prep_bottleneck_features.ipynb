```python
# Preprocessing
import os, shutil
from scipy.io import loadmat

def print_some(an_object, how_many=10):
    if type(an_object) is dict:
        how_many = min(len(an_object), how_many)
        firstpairs = {k: an_object[k] for k in sorted(an_object.keys())[:how_many]}
        print(firstpairs)

def load_dataset(path):
    data = load_files(path)
    flower_files = np.array(data['flowers'])
    flower_targets = np_utils.tocategorical(np.array(data['target']), 102)
    return flower_files, flower_targets

def flower_nr_to_file(flower_nr, flower_path=""):
    """flower filenames for numbers under 10000"""
    nr = str(flower_nr)
    lennr = len(nr)
    zeroes = (5-lennr)* "0"
    filename =  "image_"+zeroes+nr+".jpg"
    if flower_path is not "":
        filename = os.path.join(flower_path, filename)
    return filename
    

flower_image_path = os.path.join(os.getcwd(), "flowers")

# train, test, and validation datasets in .mat format
setid_file = 'setid.mat'
setid = loadmat(setid_file)
# -1 to get 0-based files
training_indexes = setid['trnid'].flatten()-1
training_files = [flower_nr_to_file(i, flower_image_path) for i in training_indexes]

validation_indexes = setid['valid'].flatten()-1
validation_files = [flower_nr_to_file(i, flower_image_path) for i in validation_indexes]

test_indexes = setid['tstid'].flatten()-1
test_files = [flower_nr_to_file(i, flower_image_path) for i in test_indexes]


# list of flowers names 
imagelabels_file = 'imagelabels.mat'
imagelabels_data = loadmat(imagelabels_file)
imagelabels = imagelabels_data['labels'].flatten()-1

train_labels = imagelabels[training_indexes]
validation_labels = imagelabels[validation_indexes]
test_labels = imagelabels[test_indexes]

# combine files and labels
train_dict = dict(zip(training_files, train_labels))
validation_dict = dict(zip(validation_files, validation_labels))
test_dict = dict(zip(test_files, test_labels))
print_some(train_dict,3)
print_some(validation_dict,3)


categories = list(range(102))
for cats in categories:
    for img_set in ["train","validation","test"]:
        directory = os.path.join(os.getcwd(),"data",img_set,str(cats))
        if os.path.exists(directory):
            pass
        else:
            os.makedirs(directory)
            print("We made", directory)
        

# copy files in dir structure
for file in train_dict.keys():    
    repl_str = "data/train/" + str(train_dict[file])
    destination = file.replace("flowers", repl_str)
    # print(file, destination)
    if not os.path.exists(destination):
        shutil.copy(file, destination)

for file in validation_dict.keys():    
    repl_str = "data/validation/" + str(validation_dict[file])
    destination = file.replace("flowers", repl_str)
    # print(file, destination)
    if not os.path.exists(destination):
        shutil.copy(file, destination)
        
for file in test_dict.keys():    
    repl_str = "data/test/" + str(test_dict[file])
    destination = file.replace("flowers", repl_str)
    # print(file, destination)
    if not os.path.exists(destination):
        shutil.copy(file, destination)
print("Done!")
 
```
